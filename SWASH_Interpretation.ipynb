# **Installation**

The latest version of the SWASH model is 7.01 and can be found [here](http://swash.sourceforge.net/). It is recommended to consult the [User Manual](http://swash.sourceforge.net/download/zip/swashuse.pdf) before running any test cases. And test cases can be found [here](http://swash.sourceforge.net/).

*Run the model*

After installation, SWASH model can be operated by window DOS commant promt. The command input file can be edited by using an ASCII editor, for example MS Notepad or [Notepad++](https://notepad-plus-plus.org/downloads/).

When the command file is ready, the model can run. Note that the command file must be save as 'filename.sws'. In DOS promt, users need to locate the directory that the command file located, and users type 'swashrun filename.sws' to run the model.

*Some tips for using DOS promt:*

```
 * cd       - change directory (or folder)
 * mkdir    - creat a new directory
 * rmdir    - remove an exsiting directory
 * dir      - list file and directories in curent directory
 * copy     - copy file(s) to other location
 * del      - delete file(s)
 * ren      - rename file or directory
 * type     - to check the content of a file

 * swashrun - run the model
 ```

# **SWASH - Input command file**


The order and list of commands that can be used with SWASH can be found in the User Manual (page 141-148).
Here is an example of a command file with the name "filename.sws" looks likes:
```

```
Note that these values are the bottom level and taken possitive downward with respect to a reference level. And the users can either

### Interpretation the input command file

As presented in the input command file above, there are serveral information needed to concern:
  * ***Heading commands***
    * **PROJ 'project_name' 'number'**: Project name and number.

  * ***Model inputs****:

    * Water level: **SET level 0.0**. It is set as 0.0 m as default.

    * Model mode: **MODE DYN ONED**: Model runs in 1D. It means that a rectilinear grid can be defined as a genuine one-dimentional one. 

    * Coordination: **COORDINATES CART**: Coordinates Cartesian. The coordinates can be set either in Cartesian or Spherial.

    * Vertical layers: **VERTICAL 2**: Vertical layers and number of layers. It is default as 01 layer.

    * Define computational grid: **CGRID REG 0. 0. 0. 40.0 0. 4000 0**: Computational grid and indicated by the command *CGRID*.
      * **REG**: The computational grid is defined as uniform and rectangular.
      * **0. 0. 0. 40.0 0. 4000 0**
        * The first "0." is the direction of the grid, which is 0.0 degrees in Cartesian convention (see more in the User Manual).
        * The next "0. 0." is the geographic location of the origin of the computational grid in the x- and y-axis, respectively (the default is 0.0, but users must give a value, even it is in default mode).
        * The "40.0 0." is the length of the computational grid in the x- and y-axis, respectively. In the example, the length of the computational grid on the x-axis is 40 m and on the y-axis is 0 m.
        * The "4000 0" is the number of grid cells on the x-axis and y-axis. So, in the example, there is 4000 grid cell in 40 m in the x-axis, and 0 grid cells in 0 m in the y-axis. 

    * Define the bottom level: **INPGRID BOTTOM REGULAR 0. 0. 0. 4000 0 0.01 0**: The bottom level can be called out by the command INPGRID. Beside bottom level, there are multiple choice for this command, for example win, friction or water level. See the User Manual for more information.
      * **REGULAR**: The input grid is uniform and rectangular.
      * **0. 0. 0. 4000 0 0.01 0**
        * The first "0." is the direction of the grid, which is 0.0 degrees in Cartesian convention (see more in the User Manual).
        * The next "0. 0." is the geographic location of the origin of the input grid in the x- and y-axis, respectively (the default is 0.0, but users must give a value, even it is in default mode).
        * The "4000 0" is the number of mesh on the x-axis and y-axis. So, in the example, there is 4000 meshes in 40 m in the x-axis, and 0 grid cells in 0 m in the y-axis.
        * The "0.01 0" is the mesh size in the x- and y-axis, respectively. In the example, the mesh size on the x-axis is 0.01 m and on the y-axis is 0 m. Note that the mesh size is often about 1/100 wavelength.

    * Reading the bathymetry: **READINP BOTTOM 1. 'profile.bot' 1 0 FREE**: The model also requires the bathymetry files in the same directory of the input command files. Using command READINP to call the bathymetry file. In the example, the file names "profile.bot".
      * **BOTTOM**: this is to define the file as bathymetry type.
      * Note that the profile file is formed as above example. It is recommended to define the number of grid cells in the profile bottom file. For example, the number of grid cells: n_{grid cell} = n_{total value of bottom file} - 1
    
    * Initial condition: **INIT ZERO** With this command, all initial water level and velocity is set to zero, so-called the cold start. In some cases, users may start a simulation from the previous run, so it is called hot start.

    * Define boundary spectra: **BOUnd SHAPespec JONswap** The "BOUND SHAPESPEC" command define the spectra in both frequency and direction at the boundary of the computational grid. The "JONSWAP" spectrum is used, and default.  

    * Boundary condition: **BOU SIDE W CCW BTYPE WEAK CON REG 0.03 1.4 0 800 20 min** \\
      * The "BOUNDCOND" (**BOU**) defines a boundary condition at the boundary. 
      * The side of the boundary is defined with "**SIDE**" command and after the direction is "**W**" (at west side), sometimes, the side can be at north side ("N").
      * **CCW BTYPE** are clockwise and are the type of boundary condition (see User Manual for more information)
      * **WEAK** the boundary condition is weakly reflection
      * **CON REG 0.03 1.4 0** this part is the wave conditions including a constant ("**CON**") and regular ("**REG**") with wave height as **0.03** m, wave period is **1.4** second. The direction of the wave is 0.0, perpendicular to the shoreline.

    * Physics conditions: These conditions include the bottom friction, wave breaking, sponger layer, and more. In the example: 
      * The friction parameter is used the Manning friction with the constant value as 0.002 as same as in the laboratory condition ("**FRIC CON**").
      * Wave breaking is set as the default unless the vertical layers are set larger than 1.0, **BREAK** is needed.
      * In some cases, waves are needed to absorb at the end of the profile to avoid unnecessary reflection, then the **SPONgelayer** is set (note that this layer is about 3-5 times the longest wavelength).
      * The non-hydrostatic and viscosity is set to default, see User Manual for more information.
      * In some other cases, the vegetation can be defined in this part. In Tung Dao thesis, the wooden fence used the vegetation commands. See User Manual. 

    * Computation schemes and numerics: See User Manual for more information. It is recommended to use as same as in the example for the wave propagation cases.
    * Time integration: **TIMEI METH**. This commnd set the model influence the time integration. There are two options can be used: "**EXPLICIT**" and "**IMPLICIT**". The explicit mode is the default, and recommended.

  * ***Model outputs***: Before extracting the outputs, it is necessary to write the computed outputs using the command "**QUANTITY**". Depending on the user's goals, the quantity can be defined by a specific command. In the example, using in the laboratory condition, and computing wave heights and set-up by the averaging surface elevation over a specific simulation time.
    * Define extract points and frames
      * **FRAME** command defines the output on a rectangular, uniform grid in a regular frame. The command **FRAME 'fr' 0. 0. 0. 40.0 0. 200 0** has the section name as "fr"; frame starts at x, y = 0; the x-length = 40 m, y-length = 0; number of points is 200 in x-direction and 0 in y-direction.
      * **POINTS** command defines a single output location. The users can define the coordinates of a single point or in a file with the same structure. **POINTS x y**
    * Writting outputs into a file: **TABLE** command can use if the output design as points, frames, curve or group (see User Manual). In the example: \\
     **TABLE 'fr' HEAD 'output_name_1.tbw' a b c** command has the writen to file command (**TABLE**); section name "fr"; writen file can have headlines (**HEAD**) or no headlines (**NOHEAD**); name of the output file "output_name_1.tbw"; and the type of outputs, a, b, and c. The options of outputs can be seen in the User Manual, page 90-100.

  * ***Computation***: 
    This is the final step to set to specify the time frame for the simulation and a time step. It is indicated by the command **COMPUTE**, the structure is the start date - time step - end time of the computation. \\
    * The start and end date has the format as default as hhmmss.000, where hh is hours, mm is minutes, and ss is seconds.
